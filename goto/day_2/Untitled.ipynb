```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.linspace(-10, 10, 1000)
```

```python
def trig(t, A, omega, phi):
    return A * np.sin(omega * t + phi)
def trig1(t, A, omega, phi):
    return A * np.cos(omega * t + phi)
```

```python
f = plt.figure(figsize=(16, 8))
plt.plot(t, trig(t, A=1, omega=1, phi=0))
plt.plot(t, trig1(t, A=1, omega=1, phi=0))
plt.show()
plt.close(f)
```
