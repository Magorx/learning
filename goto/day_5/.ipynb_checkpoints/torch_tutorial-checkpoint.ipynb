```python
import torch
import numpy as np
from matplotlib import pyplot as plt
```

```python
dots_one = np.random.normal(100, size = (100, 2), scale=20)
dots_two = np.random.normal(70, size = (100, 2), scale=20)
```

```python
dat_0 = np.concatenate((dots_one, np.ones((dots_one.shape[0], 1))), axis=1)
dat_1 = np.concatenate((dots_two, np.zeros((dots_one.shape[0], 1))), axis=1)
dat_whole = np.concatenate((dat_0, dat_1))
```

```python
plt.scatter(dat_whole[:, 0], dat_whole[:, 1], c=dat_whole[:, 2])
plt.show()
```

```python
device = torch.device('cpu')
x = torch.tensor((0, 0, 0), requires_grad=True, dtype=torch.float64, device=device)    
y = torch.tensor([137, 666, 1001], dtype=torch.float64, device=device)
```

```python
gamma = 1e-3
meter = 0
z = (x-y).pow(2).sum()
# z = torch.nn.MSELoss(y, x)
while z > 1e-6:
    z = (x-y).pow(2).sum()
    # z = torch.nn.MSELoss(x, y)
    z.backward()
    if meter % 1000 == 0:
        print('z = ', z, 'x = ', x, ' grad = ', x.grad)
    meter += 1
    with torch.no_grad():
        x -= gamma * x.grad
        x.grad.zero_()
```
